```python
from scipy.io import arff
import pandas as pd

# Se lee el archivo ARFF
data, meta = arff.loadarff('sabana_hembras_porcinas_preparado.arff')

# Se convierte a DataFrame
df = pd.DataFrame(data)

# Se decodifica todas las columnas tipo objeto
for col in df.select_dtypes([object]).columns:
    df[col] = df[col].str.decode('utf-8')

df.head(3)
```

```python
# Se eliminan las razas inválidas
df = df[~df["RAZA"].isin(["SIN_ASIGNAR"])]
df.head(3)
```

```python
# Se reemplazan por valores nulos los registros considerados atípicos o inconsistentes según límites máximos definidos para cada variable
df.loc[df["EDAD_CERDA_DIAS"] > 2500, "EDAD_CERDA_DIAS"] = None
df.loc[df["DIAS_EN_GRANJA"] > 2500, "DIAS_EN_GRANJA"] = None
df.loc[df["PESO_AL_SERVICIO"] > 300, "PESO_AL_SERVICIO"] = None
df.loc[df["DIAS_DESTETE_SERVICIO"] > 60, "DIAS_DESTETE_SERVICIO"] = None
df.loc[df["DIAS_GESTACION_ANTERIOR"] > 130, "DIAS_GESTACION_ANTERIOR"] = None
df.head(3)
```

```python
# Se crea una nueva columna que almacena la cantidad de valores nulos presentes en cada registro del DataFrame
df['total_nulos'] = df.isnull().sum(axis=1)
```

```python
# Se eliminan los registros con más del 15% de nulos (más de 2 nulos)
df = df[df['total_nulos'] <= 2]
df = df.drop(columns=['total_nulos'])
df.head()
```

```python
# Se exporta el DataFrame procesado
df.to_csv('sabana_hembras_porcinas_procesado.csv', index=False)
```
